## Gaussian Stills

This notebook produces plots of Gaussian functions that will be animated later.

First import some modules.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Now make two arrays: one with coordinates on the x-axis and the other with coordinates on the y axis.

```python
xs = np.linspace(-4, 4, num=1000)
ys = np.exp(-(xs**2/2))
```

Now plot it!

```python
fig, axs = plt.subplots(nrows=1, ncols=1)
axs.plot(xs, ys)
```
